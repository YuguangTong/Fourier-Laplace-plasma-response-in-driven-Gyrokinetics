```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import matplotlib as mpl
fs = 16
mpl.rcParams['font.size'] = fs-2
mpl.rcParams['axes.labelsize'] = fs
mpl.rcParams['axes.titlesize'] = fs
mpl.rcParams['xtick.labelsize'] = fs
mpl.rcParams['ytick.labelsize'] = fs
mpl.rcParams['legend.fontsize'] = fs-4
```

## slow mode with decorrelation rate = -0.8

```python
file_dir = '/Volumes/yuguang_sd/data/agk/lin_response/upar_160929_slow_decay/'
file, heat, t, Ebperp, Ebpar = {},  {}, {}, {}, {}
file_ind = [1, 2, 7, 8]
for i in file_ind:
    file[i] = file_dir + 'lin_upar_{}'.format(str(i)) + '.heat'
    heat[i] = np.loadtxt(file[i])
    t[i], Ebperp[i], Ebpar[i] =heat[i][:, 0], heat[i][:, 2], heat[i][:, 3]
labels = {1:'drive (decorr=0)', 2:'decay (decorr=0)', 
          3:'drive (decorr=0.1)', 4:'decay (decorr=0.1)',
          7:'drive (decorr=0.8)', 8:'decay (decorr=0.8)'}
```

```python
fig_ebpar, ax = plt.subplots(figsize=(8, 8*0.618))

for i in file_ind:
    ax.plot(t[i], Ebpar[i], label = labels[i])
ax.set_yscale('log')
# ax.set_ylim(1e-3, 1e-1)
ax.legend(loc='best')
ax.set_ylabel('$E_{B_\parallel}$')
ax.set_xlabel('$tk_\parallel v_{ti}$')
# fig_ebpar.savefig('lin_slow_decay_3.png', dpi=200)
plt.show()
```

```python
fig_mva, ax = plt.subplots(figsize=(8, 8*0.618))
for i in file_ind:
    if i % 2 == 1:
        ax.plot(t[i][1:], Ebperp[i][1:]/Ebpar[i][1:], label = labels[i])
    else:
        ax.plot(t[i], Ebperp[i]/Ebpar[i], label = labels[i])
ax.set_yscale('log')
ax.legend(loc='best')
ax.set_ylabel('MVA')
ax.set_xlabel('$tk_\parallel v_{ti}$')
# fig_mva.savefig('lin_slow_decay_mva_3.png', dpi=200)
plt.show()
```

```python
i, k = 7, 200
print(np.mean(Ebperp[i][-k:]/Ebpar[i][-k:]))
np.exp(np.mean(np.log(Ebperp[i][-k:]/Ebpar[i][-k:])))
```

```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import matplotlib as mpl
fs = 16
mpl.rcParams['font.size'] = fs-2
mpl.rcParams['axes.labelsize'] = fs
mpl.rcParams['axes.titlesize'] = fs
mpl.rcParams['xtick.labelsize'] = fs
mpl.rcParams['ytick.labelsize'] = fs
mpl.rcParams['legend.fontsize'] = fs-4
```

### Alfven and mix runs

```python
file_dir = '/Volumes/yuguang_sd/proj/new_slow_wave_cascade/stampede/low_beta/'
# file_dir = '/Volumes/yuguang_sd/data/agk/lin_response/stampede_equal_erg/'
file, heat, t, Ebperp, Ebpar = {},  {}, {}, {}, {}
filenames = [['mix_00/mix_1', 'mix_00/mix_2'], 
             ['mix_0808/mix_1', 'mix_0808/mix_2'], 
             ['alfven/alfven_1']]
for (i, fns) in enumerate(filenames):
    t[i], Ebperp[i], Ebpar[i] = np.array([]), np.array([]), np.array([])
    for fn in fns:
        file_i = file_dir + fn + '.heat'
        heat_i = np.loadtxt(file_i)
        t[i] = np.append(t[i], heat_i[:, 0])
        Ebperp[i] = np.append(Ebperp[i], heat_i[:, 2])
        Ebpar[i] = np.append(Ebpar[i], heat_i[:, 3])
```

```python
labels = {
          0:'Alfven {}'.format(r'$\overline{\omega}_0=0.9$'),
          1:'Alfven {}'.format(r'$\overline{\omega}_0=0.9$'),
          2:'Alfven {}'.format(r'$\overline{\omega}_0=0.9, DR=0.8$')}
```

```python
fig_mva, ax = plt.subplots(figsize=(8, 8*0.618))
for i in range(len(filenames)):
    ax.plot(t[i][1:], Ebperp[i][1:]/Ebpar[i][1:], label = labels[i])
ax.set_yscale('log')
ax.legend(loc='best', frameon=False)
ax.set_ylabel('MVA')
ax.set_xlabel('$tk_\parallel v_{ti}$')
# fig_mva.savefig('nl_mix_mva_1.png', dpi=200)
plt.show()
```
