```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
import matplotlib as mpl
fs = 16
mpl.rcParams['font.size'] = fs-2
mpl.rcParams['axes.labelsize'] = fs
mpl.rcParams['axes.titlesize'] = fs
mpl.rcParams['xtick.labelsize'] = fs
mpl.rcParams['ytick.labelsize'] = fs
mpl.rcParams['legend.fontsize'] = fs-4
```

```python
file_dir = '/Volumes/yuguang_sd/data/agk/lin_response/upar_160929_equal_erg/'
# file_dir = '/Volumes/yuguang_sd/data/agk/lin_response/stampede_equal_erg/'
file, heat, t, Ebperp, Ebpar = {},  {}, {}, {}, {}
file_ind = [3, 4, 5]
for i in file_ind:
    file[i] = file_dir + 'lin_upar_{}'.format(str(i)) + '.heat'
    heat[i] = np.loadtxt(file[i])
    t[i], Ebperp[i], Ebpar[i] =heat[i][:, 0], heat[i][:, 2], heat[i][:, 3]
labels = {3:'Alfven {}, slow {}'.format(r'$\overline{\omega}_0=0.9, DR=0$', r'$\overline{\omega}_0=0.13, DR=0$'),
          4:'Alfven {}, slow {}'.format(r'$\overline{\omega}_0=0.9, DR=0.8$', r'$\overline{\omega}_0=0.13, DR=0$'),
          5:'Alfven {}, slow {}'.format(r'$\overline{\omega}_0=0.9, DR=0.8$', r'$\overline{\omega}_0=0.13, DR=0.8$')}
```

```python
fig_mva, ax = plt.subplots(figsize=(8, 8*0.618))
for i in file_ind:
        ax.plot(t[i][1:], Ebperp[i][1:]/Ebpar[i][1:], label = labels[i])
ax.set_yscale('log')
ax.legend(loc='best', frameon=False)
ax.set_ylabel('MVA')
ax.set_xlabel('$tk_\parallel v_{ti}$')
# fig_mva.savefig('lin_mix_mva_1.png', dpi=200)
plt.show()
```
