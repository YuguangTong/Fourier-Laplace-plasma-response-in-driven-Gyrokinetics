```python
import numpy as np
import os
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data_dir = '/Volumes/yuguang_sd/proj/new_slow_wave_cascade/stampede/new_low_beta/'

file_names = ['alfven/alfven_large_7.heat', 'mix_080/mix_large_8.heat']

bi = 0.01
per = 2*np.pi*np.sqrt(bi)/1.0

t0 = 0
t, impci, impce, ja, ci, ce = {}, {}, {}, {}, {}, {}
for i, file in enumerate(file_names):
    path = os.path.join(data_dir, file)
    heat = np.loadtxt(path)
    impci[i] = -heat[1:, 16]
    impce[i] = -heat[1:, 17]
    ja[i] = heat[1:, 14]
    ci[i] = heat[1:, 20]
    ce[i] = heat[1:, 21]
    t[i] = heat[1:, 0]/per
    if i == len(file_names)-1:
        t0 = heat[0, 0]/per
```

```python

plt.plot(t[0], ci[0]+ ce[0])
plt.plot(t[0], -ja[0])
```
